```python
## Começando a preencher os dados com "fillna"
```

```python
df["cidade"] = df["cidade"].fillna("Desconhecida")
```

```python
df["idade"] = df["idade"].fillna(df["idade"].mean())
```

```python
df["salario"] = df["salario"].fillna(df["salario"].mean())
```
